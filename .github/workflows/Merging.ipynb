# Merging : 공통 컬럼을 찾아서 합친다

```python
import numpy as np
import pandas as pd
```

```python
frame1 = pd.DataFrame( {'id':['ball','pencil','pen','mug','ashtray'], 'price':[12.33,11.44,33.21,13.23,33.62]})
frame1
```

```python
frame2 = pd.DataFrame( {'id':['pencil','pencil','ball','pen'], 'color':['white','red','red','black']})
frame2
```

```python
pd.merge(frame1,frame2)
```

```python
frame1 = pd.DataFrame( {'id':['ball','pencil','pen','mug','ashtray'], 
                        'color':['white','red','red','black','green'],
                        'brand':['OMG','ABC','ABC','POD','POD']})
frame1
```

```python
frame2 = pd.DataFrame( {'id':['pencil','pencil','ball','pen'],
                        'brand':['OMG','POD','ABC','POD']})
frame2
```

```python
#두개의 컬럼이 같은 경우 하나를 지정해주지 않으면 합쳐지지 않음.
print(pd.merge(frame1,frame2))
```

```python
pd.merge(frame1,frame2,on='id')  #컬럼을 지정해준 경우 나머지 하나는 x,y가 붙는다
```

```python
pd.merge(frame1,frame2,on='brand')
```

```python
frame2.columns = ['sid', 'brand']
frame2
```

```python
pd.merge(frame1, frame2, left_on='id', right_on='sid')
```

```python
frame2.columns = ['brand','id']
```

```python
pd.merge(frame1,frame2,on='id')
```

```python
pd.merge(frame1,frame2,right_index=True, left_index=True)
```

```python
frame2.columns = ['brand2','id2']
```

```python
frame1.join(frame2)
```

```python
df1 = pd.DataFrame({'분기':[1,2,3,4],'팀1':np.random.randint(1,10,4)})
df1
```

```python
df2 = pd.DataFrame({'분기':[1,2,3,4],'팀2':np.random.randint(1,10,4)})
df2
```

```python
pd.merge(df1,df2,on='분기')
```

```python
pd.merge(frame1,frame2,right_index=True, left_index=True)
```

```python
df1 = pd.DataFrame({'첫번째':[1,2], '두번째':[3,4]}, index=['순서','분기'])
df1
```

```python
df2 = pd.DataFrame({'첫번째':[5,6], '두번째':[7,8]}, index=['순서','분기'])
df2
```

```python
pd.merge(df1, df2, right_index=True, left_index=True)
```

```python
df1.reset_index(inplace=True)
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1, df2], axis=1)
```
