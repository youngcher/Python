```python
index = np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
bw = 0.3
plt.axis([0,5.1,0,8])
plt.title('A Multiseries Bar Chart',fontsize=20)
plt.bar(index+bw,values1,bw,color='b')
plt.bar(index+2*bw,values2,bw,color='g')
plt.bar(index+3*bw,values3,bw,color='r')
plt.xticks(index+2*bw,['A','B','C','D','E'])
```

```python
index = np.arange(5)
values1 = [5,7,3,4,6]
values2 = [6,6,4,5,7]
values3 = [5,6,5,4,6]
bw = 0.3
plt.axis([0,8,0,5])
plt.title('A Multiseries Horizontal Bar Chart',fontsize=20)
plt.barh(index,values1,bw,color='b')
plt.barh(index+bw,values2,bw,color='g')
plt.barh(index+2*bw,values3,bw,color='r')
plt.yticks(index+0.4,['A','B','C','D','E'])
```

```python
x0 = np.arange(8)
y1 = np.array([1,3,4,6,4,3,2,1])
y2 = np.array([1,2,5,4,3,3,2,1])
plt.ylim(-7,7)
plt.bar(x0,y1,0.9,facecolor='r')
plt.bar(x0,-y2,0.9,facecolor='b')
plt.xticks(())
plt.grid(True)
for x, y in zip(x0, y1): 
    plt.text(x + 0.4, y + 0.05, '%d' % y, ha='center', va= 'bottom')
    for x, y in zip(x0, y2):
         plt.text(x + 0.4, -y - 0.05, '%d' % y, ha='center', va= 'top')
```

# grid을 활용 등치선 그리기

```python
#x,y증감 
dx = 0.01; dy = 0.01

#x와 y array
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)

#x,y 값 grid 생성
X,Y = np.meshgrid(x,y)

#메소드 생성
def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

#선그리기
C = plt.contour(X,Y,f(X,Y),8,colors='black')

plt.contourf(X,Y,f(X,Y),8)

#라벨
plt.clabel(C, inline=1, fontsize=10)
```

```python
dx = 0.01; dy = 0.01
x = np.arange(-2.0,2.0,dx)
y = np.arange(-2.0,2.0,dy)
X,Y = np.meshgrid(x,y)

def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

C = plt.contour(X,Y,f(X,Y),8,colors='black')
plt.contourf(X,Y,f(X,Y),8,cmap=plt.cm.hot)
plt.clabel(C, inline=1, fontsize=10)
plt.colorbar()
```

# 3D surface

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-2,2,0.1)
Y = np.arange(-2,2,0.1)
X,Y = np.meshgrid(X,Y)

def f(x,y):
    return (1 - y**5 + x**5)*np.exp(-x**2-y**2)

ax.plot_surface(X,Y,f(X,Y), rstride=1, cstride=1)
```

```python
from scipy.stats import norm
```

```python
x = np.arange(100)
y = np.random.normal(50, 25, 100)

plt.scatter(x, y, color='r')

below_threshold = y<33
above_threshold = np.logical_not(below_threshold) # true/false 반대로
plt.scatter(x[above_threshold], y[above_threshold], color='g') 

below_threshold2 = y<66
above_threshold2 = np.logical_not(below_threshold2)
plt.scatter(x[above_threshold2], y[above_threshold2], color='b') 

plt.show()
```

```python
x = np.arange(100)
y = np.random.normal(50,25,100)
c = [(1,0,0) if v<=33 else (0,1,0) if v>33 and v<=66 else (0,0,1) for v in y]
s = [10 if v<=33 else 20 if v>33 and v<=66 else 30 for v in y]

plt.scatter(x,y,c=c, s=s, alpha=0.7, marker="s")
plt.show()
```
